```python
#     .
def solution(distance, rocks, n):
    rocks.append(distance) #   
    rocks.sort()
    start, end = 0, distance
    answer = 0
    while start <= end:
        mid = (start+end)//2 # mid  .
        min_distance = distance
        remove_rock = 0
        current = 0 #  
        # 2, 11, 14, 17, 21, 25
        for rock in rocks:
            diff = rock - current
            if diff < mid: #    
                remove_rock += 1
            else: #    ,  
                current = rock 
                min_distance = min(min_distance, diff)
        if remove_rock> n: #    
            end = mid-1
        else:
            answer = min_distance
            start = mid+1
    return answer
```
