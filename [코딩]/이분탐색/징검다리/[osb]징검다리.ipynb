```python
# 최소값 중에 가장 큰 값.
def solution(distance, rocks, n):
    rocks.append(distance) # 마지막 거리만 추가
    rocks.sort()
    start, end = 0, distance
    answer = 0
    while start <= end:
        mid = (start+end)//2 # mid가 최소 거리.
        min_distance = distance
        remove_rock = 0
        current = 0 # 처음과의 비교
        # 2, 11, 14, 17, 21, 25
        for rock in rocks:
            diff = rock - current
            if diff < mid: # 최소 거리보다 작으면 제거
                remove_rock += 1
            else: # 최소 거리보다 크거나 같으면, 현재 
                current = rock 
                min_distance = min(min_distance, diff)
        if remove_rock> n: # 더 많이 제거한 경우
            end = mid-1
        else:
            answer = min_distance
            start = mid+1
    return answer
```
