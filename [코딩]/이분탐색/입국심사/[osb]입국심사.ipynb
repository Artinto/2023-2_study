```python
def solution(n, times):
    max_time = max(times)
    start, end = 1, n*max_time
    answer = 0
    while start<= end:
        middle = (start+end)//2 
        # 심사 가능 기준 세우기
        temp = 0
        for time in times:
            temp += middle//time
        if temp < n:
            start = middle+1
        else:
            end = middle-1
            answer = middle
    return answer
```
