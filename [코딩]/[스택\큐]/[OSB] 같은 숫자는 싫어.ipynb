* 굳이 기존의 배열의 전위치를 찾는게 아니라 ANSWER의 배열의 마지막만 비교해도 된다.

```python
def solution(arr):
    answer = []
    for v in (arr):
        if answer[-1:] == [v]: # 기본적으로 answer[-1]은 빈 리스트일때 실행 안되지만, :를 통해 실행하면 빈 리스트여도 실행가능. 그리고 return 시 리스트 형태로 반환 타라서 v 대신 [v] 모양 사용.
            continue
        else:
            answer.append(v)
    return answer
```
