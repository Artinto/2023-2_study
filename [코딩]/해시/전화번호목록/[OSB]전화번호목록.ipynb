* deque를 이용해서 시간복잡도를 낮춰보기

```python
from collections import deque
def solution(phone_book):
    d = deque(phone_book)
    len_d = len(d)
    for _ in range(len_d):
        a = d.popleft()
        length_a = len(a)
        for v in d:
            length_v = len(v)
            if a == v[:length_a] or v == a[:length_v] :
                answer = False
                return answer
            else:
                answer = True
    return answer
```

* zip 함수를 이용해서 시간 복잡도 낮추기

```python
def solution(phone_book):
    d = list(set(phone_book))
    for v1, v2 in zip(d, d[1:]):
        len_v1, len_v2 = len(v1), len(v2)
        if v2 == v1[:len_v2] or v1 == v2[:len_v1] :
            answer = False
            return answer
        else:
            answer = True
    return answer
```

* 결국 정렬되어 있으므로 연달아 있는 2개만 확인하면됨.

```python
def solution(phone_book):
    d = sorted(phone_book)
    answer = True
    for i in range(len(d)- 1):
        v1, v2 = d[i], d[i+1]
        if len(v2)> len(v1):
            if v1 == v2[:len(v1)]:
                answer = False
                break
    return answer
```
